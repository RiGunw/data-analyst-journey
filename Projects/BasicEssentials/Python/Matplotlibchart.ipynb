```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/spam.csv', sep=',', header=0, encoding='latin1')
```

```python
df
```

```python
df_salary_mean = df["salary"].mean()
print(df_salary_mean)
```

```python
result = df.loc[df['salary'] > df_salary_mean]
print(result)
```

```python
x = df["id"]
y = df["salary"]

plt.bar(x, y, color='violet')
plt.title('Chart Title')
plt.xlabel('ID')
plt.ylabel('Salary')
plt.axhline(df_salary_mean)
plt.show()
```

```python
x = df["id"]
y = df["salary"]

plt.plot(x, y, linestyle = 'dotted')
plt.title('Chart Title')
plt.xlabel('ID')
plt.ylabel('Salary')
plt.axhline(df_salary_mean)
plt.show()
```
